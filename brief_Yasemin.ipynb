```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

data = pd.read_csv('Data/Billionaires Statistics Dataset.csv')
data.head(2)
```

```python
data.info()
```

```python
top_countries = data["country"].value_counts().head(10)
top_countries
```

```python
plt.figure(figsize=(8, 4))

top_countries.plot(kind='bar', color = "purple")
plt.xlabel("Pays")
plt.ylabel("Nombre de milliardaires")
plt.title("Top 10 des Pays avec le plus de Milliardaires")
plt.xticks(rotation=75)

plt.show()
```

```python
df = pd.DataFrame(data)

# Utilisez pandas pour transformer le DataFrame en un format adaptÃ© pour le graphique
df_melted = df.melt('country', var_name='gender', value_name='count')

# CrÃ©ez le graphique catplot avec deux barres pour chaque pays
sns.catplot(x='country', y='count', hue='gender', kind='bar', data=df_melted, height=6, aspect=2)

# Ajoutez un titre et des Ã©tiquettes d'axe
plt.title('Proportion d\'hommes et de femmes par pays')
plt.xlabel('Pays')
plt.ylabel('Nombre de milliardaires')

# Affichez le graphique
plt.show()
```
