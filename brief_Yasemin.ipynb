```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

data = pd.read_csv('Data/Billionaires Statistics Dataset.csv')
data.head(2)
```

```python
data.info()
```

```python
top_countries = data["country"].value_counts().head(10)
top_countries
```

```python
plt.figure(figsize=(8, 4))

top_countries.plot(kind='bar', color = "purple")
plt.xlabel("Pays")
plt.ylabel("Nombre de milliardaires")
plt.title("Top 10 des Pays avec le plus de Milliardaires")
plt.xticks(rotation=75)

plt.show()
```

```python
df = pd.DataFrame(data.loc[data.country.isin(top_countries.index)])

__ = df.groupby(['country','gender'])['personName'].count().sort_values(ascending=False)

__.index[0][0]##.loc[__.country.isin(top_countries)]
```

```python
df = pd.DataFrame(data)


df_melted = df.melt('country', var_name='gender', value_name='count')
sns.catplot(x='country', y='count', hue='gender', kind='bar', data=df_melted, height=6, aspect=2)

plt.title('Proportion d\'hommes et de femmes par pays')
plt.xlabel('Pays')
plt.ylabel('Nombre de milliardaires')


plt.show()
```
